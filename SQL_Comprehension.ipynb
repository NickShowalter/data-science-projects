```python
Run Saver
/* 
These scripts were created with the intent to reduce time doing manual Gap Analysis work.
All scripts will need to be updated before each use following the comments located to the right of each script.
In this User Friendly Version all specific database references are made generic. i.e. DB1, DB2, a, b...
User only has to replace the indicated areas which are noted
Output times will vary with connection and size of data being used
Personal longest output time: 22 seconds (If longer than that consider checking connection and script)
***Feel free to save this to your own device. Feel free to alter, add, take away any scripts you feel necessary. Alex Freberg will have original if there are any mistakes on your local copy in the future.***
*VERSION HISTORY*
Created	FULL GAP ANALYSIS	:                              2019.08.08
Updated FULL GAP ANALYSIS to work out visible flaws	:      2019.08.13
Added COMPARING A TABLE TO A TABLE   :                     2019.09.19
Added COMPARING DATA TO DATA     :                         2019.09.20
Added COMPARING 2 DATA TO DATA     :                       2019.09.25
*/


----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

-- COMPARING ALL FIELDS AND COLUMNS (FULL GAP ANALYSIS)



select isnull(Database1.table_name, Database2.table_name) as [Database 1 Table], Database1.column_name as [Database 1 Column], 
CASE WHEN Database1.column_name = DataBase2.column_name
    THEN 'Match'
    ELSE NULL
END as [Match between Tables],
isnull(DataBase2.table_name, DataBase1.table_name) as [Database 1 Table], DataBase2.column_name as [Database 2 Column]

from
(select DB1Tab.name as table_name, DB1Col.name as column_name
   from [Database1].sys.tables as DB1Tab                      -- Only have to replace Database name in this Select statement (of 1st database)
        inner join [Database1].sys.columns as DB1Col
            on DB1Tab.object_id = DB1Col.object_id) Database1

full outer join

(select DB2Tab.name as table_name, DB2Col.name as column_name
   from [Database2].sys.tables as DB2Tab                 -- Replace Database name in this Select statement(of 2nd database)
        inner join [Database2].sys.columns as DB2Col
        on DB2Tab.object_id = DB2Col.object_id) DataBase2

on Database1.table_name = DataBase2.table_name 
and Database1.column_name = DataBase2.column_name
order by 1,2,5





-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

-- COMPARING A TABLE TO A TABLE

-- This is unique because the databases nor the tables need to be the same
-- It will only be comparing the column names to eachother


WITH CTE (
[Database 1 Column]
,[Match between Tables]
,[Database 2 Column]
)
AS
(
select [Database1].column_name as [Database 1 Column], 
CASE WHEN [Database1].column_name = [DataBase2].column_name
    THEN 'Match'
    ELSE NULL
END as [Match between Tables],
 [DataBase2].column_name as [Database 2 Column]

from
(select DB1Tab.name as table_name, DB1Col.name as column_name
   from [Database1].sys.tables as DB1Tab                      --Only have to replace Database name in this Select statement (of 1st database)
        inner join [Database1].sys.columns as DB1Col
            on DB1Tab.object_id = DB1Col.object_id
	where DB1Tab.name = 'Rx_20161004') Database1   --insert table from  First Database

full outer join

(select DB2Tab.name as table_name, DB2Col.name as column_name
   from [Database2].sys.tables as DB2Tab                 --and replace Database name in this Select statement(of 2nd database)
        inner join [Database2].sys.columns as DB2Col
        on DB2Tab.object_id = DB2Col.object_id
	where DB2Tab.name = 'Rx') DataBase2    --insert table from Second Database

on Database1.column_name = DataBase2.column_name
)
Select *
From CTE
order by 3,2,1




-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------


-- COMPARING UNIQUE DATA FROM COLUMNS OF ONE TABLE TO ANOTHER (COMPARING DATA TO DATA)


WITH CTE (
[Table 1 Name]
,[Column 1 Name]
,[Table 1 Data Point]
,[Match between Tables]
,[Table 2 Name]
,[Column 2 Name]
,[Table 2 Data Point]
)
AS (

SELECT distinct
'Custom_Column1'                                                 --Manually input First Table Name
,'Custom_Column2'                                                          --Manually input First Column Name
,a.column_name1                                                          --Update to First column that will be compared to the Second Column
,CASE WHEN a.column_name1 = b.column_name2                  -- Both need to be updated with First and Second column names
    THEN 'Match'
    ELSE NULL
END
,'Custom_Column1'                                                --Manually input First Table Name
,'Custom_Column2'                                                          --Manually input First Column Name
,b.column_name1                                                          --Update to First column that will be compared to the Second Column
FROM [Database1].dbo.[Table1] a                          -- Update to First Database and Table 
	full outer join [Database2].dbo.[Table2] b      -- Update to Second Database and Table 
		ON a.column_name1 = b.column_name2                      --Update to the columns that you will be comparing - making the columns equal to each other ensures an accurate output   
)
Select *
From CTE
--where [Match between Tables] = 'Match'
Order by 4 desc, 3






-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

-- COMPARING TWO UNIQUE DATA FROM COLUMNS OF ONE TABLE TO ANOTHER (COMPARING 2 DATA TO DATA)

--This can be added and built on to compare as many data point/columns as are in a table, but must be manually updated with each additional data point


WITH CTE (
[Table 1 Name]
,[Column 1 Name]
,[Table 1 Data Point 1]
,[Table 1 Data Point 2]
,[Match between Tables]
,[Table 2 Name]
,[Column 2 Name]
,[Table 2 Data Point 1]
,[Table 2 Data Point 2]
)
AS (

SELECT distinct
'rep_reports'                                                 --Manually input First Table Name
,'primary_param_nm'                                                          --Manually input First Column Name
,a.column_name1
,a.column_name2                                                          --Update to First column that will be compared to the Second Column
,CASE WHEN a.column_name1 = b.column_name2                  -- Both need to be updated with First and Second column names
	  AND a.column_name2 = b.column_name3                 -- Both need to be updated with First and Second column names
    THEN 'Match'
    ELSE NULL
END
,'rep_reports'                                                --Manually input First Table Name
,'primary_param_nm'                                                          --Manually input First Column Name
,b.column_name1
,b.column_name2                                                          --Update to First column that will be compared to the Second Column

FROM [Database1].dbo.[Table1] a                          -- Update to First Database and Table 
	full outer join [Database2].dbo.[Table2] b      -- Update to Second Database and Table 
		ON a.column_name1 = b.column_name2                      --Update to the columns that you will be comparing - making the columns equal to each other ensures an accurate output 
		and a.column_name2 = b.column_name3                   --Update to the columns that you will be comparing - making the columns equal to each other ensures an accurate output 
)
Select *
From CTE
--where [Match between Tables] = 'Match'
Order by 5 desc, 3
```
