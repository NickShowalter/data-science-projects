```python
import pandas as pd
import dtale
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
run = pd.read_csv('Running_Data.csv')
```

```python
run.dropna()
```

```python
run.drop('Unnamed: 10', axis=1, inplace=True)
```

```python
dtale.show(run, ignore_duplicate=True)
```

```python
run.dtypes
```

```python
num_list = run["Distance(mi)"]
float_list = [float(x) for x in num_list]
print(num_list)
```

```python
run = run[pd.notnull(run['Distance(mi)'])]
```

```python
run.dropna(subset = ['Distance(mi)'])
run.dropna(subset = ['Max Speed mph'])
```

```python
distance = run['Distance(mi)']
speed = run['Max Speed mph']
x = distance.values.reshape(-1,1)
y = speed.values.reshape(-1,1)
reg = LinearRegression()
reg.fit(x,y)

pred = reg.predict(x[[1]])
pred
```

```python
plt.scatter(x,y, color='blue')
plt.plot(x, reg.predict(x))
```
